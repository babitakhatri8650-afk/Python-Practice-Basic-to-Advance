```python
# 14
rows = 5   # half height

# upper half
for i in range(1, rows + 1):
    for j in range(rows - i):
        print(" ", end=" ")
    for k in range(2 * i - 1):
        print("*", end=" ")
    print()

# lower half
for i in range(rows - 1, 0, -1):
    for j in range(rows - i):
        print(" ", end=" ")
    for k in range(2 * i - 1):
        print("*", end=" ")
    print()
    
```

```python
# Q 15
rows = 7

for i in range(rows):
    for j in range(5):
        if (
            (j == 0 or j == 4) and i != 0 or
            (i == 0 and j != 0 and j != 4) or
            (i == 3)
        ):
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
