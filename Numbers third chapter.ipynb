```python
# Q.10
angle=float(input("Enter the angle in degree:"))
import math
radian=math.radians(angle)
print(f"The angle of sin is {math.sin(radian)}")
print(f"The angle of cos is {math.cos(radian)}")
print(f"The angle of tan  is {math.tan(radian)}")
```
